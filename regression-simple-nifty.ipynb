```python
#Simple Linear Regression Model that predicts Close value for the day
#given the Open,High & Low for the Day. Implemented using Normal Equation.



import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory


f=pd.read_csv('/kaggle/input/stock-market-index-data-india-1990-2022/NIFTY 50_Data.csv')
f=f[['Open','High','Low','Close']]
f.insert(0,"Bias",[1 for i in range(len(f['Open']))]);
train=f[0:1000]
train_poly_var=train['High']*train['Low']*0.1;
train=train[['Bias','Open','High','Low','Close']]
train.insert(2,"HiLo",train_poly_var);

test=f[1000:1100]
test_poly_var=test['High']*test['Low']*0.1;
test=test[['Bias','Open','High','Low','Close']]
test.insert(2,"HiLo",test_poly_var);
test_in=test[['Bias','Open','High','Low','HiLo']]
test_op=test[['Close']]

x_train = train[['Bias','Open','HiLo','High','Low']];
y_train=train[['Close']]

x_train_transpose=x_train.transpose();
X=x_train_transpose.dot(x_train);
X=pd.DataFrame(np.linalg.pinv(X.values), X.columns, X.index)
#tt=x_train_transpose.dot(y_train);
w=X.dot(x_train_transpose);
w=w.dot(y_train);

test_d=test_in[0:1]
pred=test_in.dot(w);

error=abs(test_op-pred);
print("Predicted Value \n",pred);

print("Actual Value \n",test_op);

print("Error for each Value \n",error);

print("Max Error Value \n",error.max());

"""
#custom value bias=1
OPen
18,094.75
High
18,134.75
Low
18,000.65
Prev close
18,015.85 

test_d=pd.DataFrame(np.array([[1,18094.75,18134.75,18000.65]]),
                    columns = ['Bias','Open','High','Low'])
pred=test_d.dot(w);
print(pred)

"""





```
